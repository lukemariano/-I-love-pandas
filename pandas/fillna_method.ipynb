## **Fill in Missing DataFrame Values with the fillna Method**

```python
import pandas as pd
```

```python
nba = pd.read_csv("nba.csv")
nba.tail()
```

```python
nba.fillna(0)
```

```python
nba['College'].fillna('Unknown')
```

Com o comando acima estamos criando uma cópia da série original do dataframe, e não alterando nada no nosso dataframe inicial.  

Atribuir essa alteração a uma variável é uma boa prática uma vez que permite manter explicítas as nossas regras de negócio.  

Uma outra alternativa para realizar alterações em um dataframe sem alterar a fonte de dados original seria simplesmente utilizando o método 'copy', que na realidade é o que acontece sobre o capô do pandas quando utilizamos o parâmetro 'inplace=True';

### Qual seria a maneira recomendada de realizarmos alterações na fonte de dados que estamos trabalhando sem utilizar o método inplace?

```python
nba['College'] = nba['College'].fillna('Unknown') # Essa seria a forma correta
nba.tail()
```

```python
nba['Salary'] = nba['Salary'].fillna(value = 0)
nba.tail()
```
