## SORT A DATAFRAME VALUES WITH A SORT METHOD 1

```python
import pandas as pd
```

```python
nba = pd.read_csv("nba.csv")
nba.tail()
```

```python
nba["Name"].sort_values() # funciona sem par√¢metros para objetos series
```

```python
nba.sort_values(by = "Name") # precisa especificar os par√¢metros corretamente para um objeto dataframe
```

```python
nba.sort_values("Salary", na_position="first")
```

```python
# manipulando vari√°veis com regras de neg√≥cio

nba_sorted = nba.sort_values(by = "Salary", ascending = False)
nba_sorted.head()
```

# **Sort a Dataframe values with sort_values method II**

```python
nba = pd.read_csv("nba.csv").dropna(how = "all")
nba.tail()
```

```python
nba.sort_values(by = ["Team", "Name"], ascending = [True, False])
```
