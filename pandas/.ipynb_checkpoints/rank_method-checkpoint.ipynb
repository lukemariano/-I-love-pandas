# **Rank values with the rank Method**

```python
import pandas as pd
```

```python
nba = pd.read_csv("nba.csv").dropna(how = "all")
nba.tail()
```

```python
nba['Salary'] = nba['Salary'].fillna(0)
```

```python
nba.head()
```

```python
nba['Salary Rank'] = nba['Salary'].rank(ascending = False).astype('int')
```

```python
nba.sort_values(by = 'Salary Rank')
```
