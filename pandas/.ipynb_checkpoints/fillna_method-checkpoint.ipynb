## **Fill in Missing DataFrame Values with the fillna Method**

```python
import pandas as pd
```

```python
nba = pd.read_csv("nba.csv")
nba.tail()
```

```python
nba.fillna(0)
```

```python
nba['College'].fillna('Unknown')
```

Atribuir essa alteraÃ§Ã£o a uma variÃ¡vel Ã© uma boa prÃ¡tica uma vez que permite manter explicÃ­tas as nossas regras de negÃ³cio.

### Qual seria a maneira recomendada de realizarmos alteraÃ§Ãµes na fonte de dados que estamos trabalhando sem utilizar o mÃ©todo inplace?

```python
nba['College'] = nba['College'].fillna('Unknown') # Essa seria a forma correta
nba.tail()
```

```python
nba['Salary'] = nba['Salary'].fillna(value = 0)
nba.tail()
```
