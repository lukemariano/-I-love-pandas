## **The astype Method**

```python
import pandas as pd
```

```python
nba = pd.read_csv("nba.csv").dropna(how = "all")
nba.tail()
```

O **ponto número um** que precisamos destacar aqui é que sempre que tivermos valores ausentes em uma coluna com valores numéricos de um dataset, o pandas irá importar esses valores como 'float' para suportar essas extremidades de valores ausentes. Não sei bem o porquê do pandas utilizar essa técnica mas se estão utilizando é porque isso foi visto como algo bom durante o processo de design da biblioteca.

```python
print(f"The column have any null value? --> {nba['Age'].hasnans}")
nba['Age'].astype("int") # essa é a abordagem mais segura para conversões utilizando o método 'astype'
nba['Age'].astype(int) # se o tipo de dados existir em python, podemos passar o construtor como parâmetro para o dtype.
```

É importante verificar se a coluna que queremos alterar o tipo de valor possui valores nulos, pois se tiver, existe um potencial muito grande de qualquer operação de conversão de valores falhar.

```python
nba['Age'] = nba['Age'].astype("int")
```

```python
nba.tail(n = 6)
```

```python
print(f"The column have any null value? --> {nba['Salary'].hasnans}")
nba['Salary'] = nba['Salary'].fillna(0)
print(f"The column have any null value after apply fillna method? --> {nba['Salary'].hasnans}")
nba.tail()
```
